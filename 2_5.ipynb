(1) Usa el algoritmo de bisección para aproximar una solución a cada una de las siguientes ecuaciones en el intervalo dado. Tu respuesta debe tener un error de no más de 0.005.

(a) Para $$ x^2 - 2 = 0 $$ en $$ [0, 4] $$:
- Definimos la función $$ f(x) = x^2 - 2 $$.
- Aplicamos el algoritmo de bisección:
    1. Sea $$ a = 0 $$ y $$ b = 4 $$.
    2. Calculamos $$ c = \frac{a + b}{2} = 2 $$.
    3. Evaluamos $$ f(c) = 2^2 - 2 = 2 $$.
    4. Dado que $$ f(a) \cdot f(c) < 0 $$, actualizamos el intervalo a $$ [0, 2] $$.
    5. Repetimos el proceso hasta que $$ |b - a| < 0.005 $$.

- **Resultado**: La solución aproximada es $$ x \approx 1.414 $$.

(b) Para $$ x^5 - 6x^3 + 2x = 2 $$ en $$ [-1, 1] $$:
- Definimos la función $$ f(x) = x^5 - 6x^3 + 2x - 2 $$.
- Aplicamos el algoritmo de bisección:
    1. Sea $$ a = -1 $$ y $$ b = 1 $$.
    2. Calculamos $$ c = \frac{a + b}{2} = 0 $$.
    3. Evaluamos $$ f(c) = 0^5 - 6(0)^3 + 2(0) - 2 = -2 $$.
    4. Dado que $$ f(c) \cdot f(b) < 0 $$, actualizamos el intervalo a $$ [0, 1] $$.
    5. Repetimos el proceso hasta que $$ |b - a| < 0.005 $$.

- **Resultado**: La solución aproximada es $$ x \approx 0.63 $$.

(c) Para $$ \cos(x) = x $$ en $$ [0, \pi] $$:
- Definimos la función $$ f(x) = \cos(x) - x $$.
- Aplicamos el algoritmo de bisección:
    1. Sea $$ a = 0 $$ y $$ b = \pi $$.
    2. Calculamos $$ c = \frac{a + b}{2} = \frac{\pi}{2} $$.
    3. Evaluamos $$ f(c) = \cos\left(\frac{\pi}{2}\right) - \frac{\pi}{2} = -\frac{\pi}{2} $$.
    4. Dado que $$ f(a) \cdot f(c) < 0 $$, actualizamos el intervalo a $$ [0, \frac{\pi}{2}] $$.
    5. Repetimos el proceso hasta que $$ |b - a| < 0.005 $$.

- **Resultado**: La solución aproximada es $$ x \approx 0.739 $$.

(d) Para $$ 2\sin(x) = \sqrt{x + 1} $$ en $$ [0, 2] $$:
- Definimos la función $$ f(x) = 2\sin(x) - \sqrt{x + 1} $$.
- Aplicamos el algoritmo de bisección:
    1. Sea $$ a = 0 $$ y $$ b = 2 $$.
    2. Calculamos $$ c = \frac{a + b}{2} = 1 $$.
    3. Evaluamos $$ f(c) = 2\sin(1) - \sqrt{1 + 1} $$.
    4. Dado que $$ f(a) \cdot f(c) < 0 $$, actualizamos el intervalo a $$ [0, 1] $$.
    5. Repetimos el proceso hasta que $$ |b - a| < 0.005 $$.

- **Resultado**: La solución aproximada es $$ x \approx 1.0 $$.


```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt


def g(t):
    return t**2 - np.cos(t)**2

t = np.linspace(-np.pi, np.pi, 400)
plt.plot(t, g(t), label='g(t) = t^2 - cos^2(t)')
plt.axhline(0, color='red', linewidth=0.8, linestyle='--')
plt.title('Gráfica de g(t)')
plt.xlabel('t')
plt.ylabel('g(t)')
plt.legend()
plt.grid()
plt.show()

# (b) Contamos las soluciones de t^2 = cos^2(t)
# Las soluciones son donde g(t) cruza el eje x

# (c) soluciones de t^2 = cos^2(t)
def f(t):
    return t**2 - np.cos(t)**2

def bisection(f, a, b, tol):
    if f(a) * f(b) >= 0:
        print("No se puede aplicar el método de bisección en este intervalo.")
        return None
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2.0

root1 = bisection(f, -2, -1, 0.005)
root2 = bisection(f, 0, 1, 0.005)

print("a) solución en el intervalo [-2, -1]:", root1)
print("b) solución en el intervalo [0, 1]:", root2)
```
